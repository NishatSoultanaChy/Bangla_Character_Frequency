```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
def create_dictionary2(txt):
    dictionary = {}
    i=0
    for x in set(txt):
        dictionary[x] = txt.count(x)/len(txt)
    return dictionary
```

```python
test = open("BanglaText.txt",encoding='utf-8').read()
```

```python
test_dict = create_dictionary2(test)

plt.title('Bangla letter counter')
plt.ylabel('Frequency')
plt.xlabel('Character')
y_pos = np.arange(len(test_dict.keys()))
plt.figure(figsize=(23, 6)) 
#prop = FontProperties()
#prop.set_file('kalpurush.ttf')
prop = fm.FontProperties(fname='kalpurush.ttf')
plt.xticks(y_pos, test_dict.keys(), fontproperties=prop, fontsize=15)

#plt.bar(y_pos, test_dict.values(), align='center', alpha=0.5)

#plt.show() 

plt.bar(test_dict.keys(), test_dict.values(),width=.5, color='g') 
```
